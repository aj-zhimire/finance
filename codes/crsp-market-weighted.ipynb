```python
import wrds as wrds
import pandas as pd

db = wrds.Connection()

```

```python
# df_trade daily + CRSP market return

query = """
SELECT 
    d.date as trade_date,
    d.permno,
    n.ticker,
    d.prc AS price,
    d.ret,
    d.vol,
    d.shrout,
    m.vwretd as mkt_ret,
    d.ret - m.vwretd as excess_ret
FROM crsp.dsf d
JOIN crsp.dsenames n
    ON d.permno = n.permno
JOIN crsp.dsi m
    ON d.date = m.date
WHERE n.ticker IN ('ARCC','MAIN')
  AND d.date BETWEEN n.namedt AND n.nameendt
  AND d.date >= '2015-12-01'
ORDER BY d.permno, d.date
"""

df_market_return = db.raw_sql(query)
df_market_return.head()
df_market_return.to_csv("../data/crsp_market_return.csv", index=False)
```
