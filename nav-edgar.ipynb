```python
import requests
import pandas as pd

headers = {'User-Agent': 'pushpa.sharma@mavs.uta.edu'}
cik = '0001287750'

url = f"https://data.sec.gov/api/xbrl/companyconcept/CIK{cik}/us-gaap/NetAssetValuePerShare.json"

response = requests.get(url, headers=headers)

if response.status_code == 200:
    data = response.json()

    df = pd.DataFrame(data['units']['USD/shares']).copy()

    # Clean dates
    df.loc[:, 'end'] = pd.to_datetime(df['end'], errors='coerce')
    df.loc[:, 'filed'] = pd.to_datetime(df['filed'], errors='coerce')

    df = df.dropna(subset=['end'])

    # Filter from June 2015
    df = df[df['end'] >= pd.Timestamp('2015-06-01')]

    # Keep filings only
    df = df[df['form'].isin(['10-Q', '10-K'])]

    # Deduplicate
    df = df.sort_values('end').drop_duplicates(subset=['end'], keep='last')

    print(df[['end', 'val', 'form', 'frame']].tail(10))

else:
    print(f"Error: {response.status_code}")
```
